```python
# Iplorting Python Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("customer_retail csv file.csv")
```

```python
df = df.dropna()
```

```python
df = df[["Quantity", "UnitPrice", "Country"]]
```

```python
encoder = LabelEncoder()
df["Country_encoded"] = encoder.fit_transform(df["Country"])
```

```python
plt.scatter(df["Quantity"], df["UnitPrice"])
plt.xlabel("Quantity")
plt.ylabel("UnitPrice")
plt.title("Scatter Plot of Quantity vs UnitPrice")
plt.show()
```

```python
x = df[["Quantity", "UnitPrice"]]
y = df["Country_encoded"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print("Logistic Regression Model")
# Defining the Logistic Regression Model
logistic_model = LogisticRegression()
# Fitting the Pattern
logistic_model.fit(x_train, y_train)
y_pred_logistic = logistic_model.predict(x_test)
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)
print("Accuracy:", accuracy_logistic)

confusion_matrix_logistic = confusion_matrix(y_test, y_pred_logistic)
print("Confusion Matrix:")
print(confusion_matrix_logistic)
```

```python
print("Decision Tree Model")
# Defining the Decision Tree Model
decision_tree_model = DecisionTreeClassifier()
# Fitting the Pattern
decision_tree_model.fit(x_train, y_train)
y_pred_decision_tree = decision_tree_model.predict(x_test)
accuracy_decision_tree = accuracy_score(y_test, y_pred_decision_tree)
print("Accuracy:", accuracy_decision_tree)

confusion_matrix_decision_tree = confusion_matrix(y_test, y_pred_decision_tree)
print("Confusion Matrix:")
print(confusion_matrix_decision_tree)
```

```python
print("knn Model")
# Defining the knn Model
knn_model = KNeighborsClassifier(n_neighbors=3)
# Fitting the Pattern
knn_model.fit(x_train, y_train)
y_pred_knn = knn_model.predict(x_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("Accuracy:", accuracy_knn)

confusion_matrix_knn = confusion_matrix(y_test, y_pred_knn)
print("Confusion Matrix:")
print(confusion_matrix_knn)
```

```python
models = ["Logistic Regression",
"Decision Tree",
"KNN"
]
accuracy = [ accuracy_logistic,
accuracy_decision_tree,
accuracy_knn
]
```

```python
plt.bar(models, accuracy)
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison")
plt.show()
```
